```python
import pinecone
from pinecone import Pinecone, ServerlessSpec
import os
from dotenv import load_dotenv, find_dotenv
```

```python
load_dotenv(find_dotenv(), override = True)
```

```python
# pc = Pinecone(api_key = os.environ.get("PINECONE_API_KEY"))
pc = Pinecone(api_key=os.getenv("PINECONE_API_KEY"))
```

```python
pc.list_indexes()
```

```python
index_name = "my-index"
dimension = 3
metric = "cosine"
```

```python
# if index exist, delete it
if index_name in [index.name for index in pc.list_indexes()]:
    pc.delete_index(index_name)
    print(f"{index_name} succesfully deleted.")
else:
     print(f"{index_name} not in index list.")
```

```python
pc.list_indexes()
```

```python
# create an index
pc.create_index(
    name = index_name, 
    dimension = dimension, 
    metric = metric, 
    spec = ServerlessSpec(
        cloud = "aws", 
        region = "us-east-1")
    )
```

```python
pc.list_indexes()
```

```python
index_name_2 = "my-index-2"
dimension_2 = 1536
metric_2 = "euclidean"
```

```python
if index_name_2 in [index.name for index in pc.list_indexes()]:
    pc.delete_index(index_name_2)
    print(f"{index_name_2} succesfully deleted.")
else:
    print(f"{index_name_2} not in index list.")
```

```python
pc.create_index(
    name=index_name_2,
    dimension=dimension_2,
    metric=metric_2,
    spec=ServerlessSpec(
        cloud='aws',
        region='us-east-1'
    )
)
```

```python
pc.list_indexes()
```

```python
index = pc.Index(name = index_name)
```

```python
# [legs,wings,tails]
index.upsert([
        ("Dog",[4.,0.,1.]),
        ("Cat",[4.,0.,1.]),
        ("Chicken",[2.,2.,1.]),
        ("Mantis",[6.,2.,3.]),
        ("Elephant",[4.,0.,1.])
])
```
