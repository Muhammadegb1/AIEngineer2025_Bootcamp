```python
from langchain_openai.chat_models import ChatOpenAI
```

```python
chat = ChatOpenAI(model_name = 'gpt-4', 
                  model_kwargs = {'seed':365}, 
                  temperature = 0,
                 max_tokens = 100)
```

```python
response = chat.invoke(''' I've recently adopted a dog. 
                       Could you suggest some dog names? ''')
```
