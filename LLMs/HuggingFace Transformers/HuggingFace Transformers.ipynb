## Pre-trained Tokenizers

```python
from transformers import AutoTokenizer
```

```python
model = "bert-base-uncased"
```

```python
tokenizer = AutoTokenizer.from_pretrained(model)
```

```python
sentence = "I'm so excited to be learning about large language models"
```

```python
# input_ids: numerical IDs representing the tokens in the sentence
# token_type_ids: indicate which sentence each token belongs to (all 0 for a single sentence)
# attention_mask: marks real tokens (1) and padding tokens (0)

#101 = [CLS] (marks the beginning of a sentence)
#102 = [SEP] (marks the end of a sentence)

input_ids = tokenizer(sentence)
print(input_ids)
```

```python
tokens = tokenizer.tokenize(sentence)
```

```python
print(tokens)
```

```python
token_ids = tokenizer.convert_tokens_to_ids(tokens)
```

```python
print(token_ids)
```

```python
decoded_ids = tokenizer.decode(token_ids)
print(decoded_ids)
```

```python
tokenizer.decode(101)
```

```python
tokenizer.decode(102)
```

```python
# XLNet tokenization example
# Use the XLNet tokenizer to split a sentence into subword tokens,
# convert them to token IDs, and inspect special tokens (<cls>, <sep>)

# cls - classification token
# sep - separator token

model2 = "xlnet-base-cased"
```

```python
tokenizer2 = AutoTokenizer.from_pretrained(model2)
```

```python
input_ids = tokenizer2(sentence)
```

```python
print(input_ids)
```

```python
# View how the sentence is split into tokens

tokens = tokenizer2.tokenize(sentence)
print(tokens)
```

```python
token_ids = tokenizer2.convert_tokens_to_ids(tokens)
print(token_ids)
```

```python
tokenizer2.decode(4)
```

```python
tokenizer2.decode(3)
```

```python
decoded_idss2 = tokenizer2.decode(token_ids)
print(decoded_idss2)
```

## Huggingface and Pytorch/Tensorflow

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch
```

```python
print(sentence)
print(input_ids)
```

```python
tokenizer = AutoTokenizer.from_pretrained("distilbert-base-uncased-finetuned-sst-2-english")
```

```python
# Convert sentence to PyTorch tensors for the model

input_ids_pt = tokenizer(sentence, return_tensors ="pt")
print(input_ids_pt)
```

```python
model = AutoModelForSequenceClassification.from_pretrained("distilbert-base-uncased-finetuned-sst-2-english")
```

```python
# Compute predictions without gradients, pick the class with highest score, and get its label
# # The '**' unpacks a dictionary into keyword arguments for the function. input_ids' and 'attention_mask'
# == model(input_ids=input_ids_pt['input_ids'], attention_mask=input_ids_pt['attention_mask'])

with torch.no_grad():
    logits = model(**input_ids_pt).logits
    
# # logits.argmax() → returns the position of the max value in logits
# .item() → converts the single-value tensor to a regular Python integer
predicted_class_id = logits.argmax().item()
model.config.id2label[predicted_class_id]
```
