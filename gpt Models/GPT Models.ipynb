## Generating Text

```python
# response = openai.Completion.create(...) כאן אתה שולח בקשה לשרת של OpenAI.
# engine="davinci-002" → שם המודל (ישן, כבר לא קיים).
# prompt → הטקסט שאתה שולח למודל.
# max_tokens=10 → כמה טקסט המודל יכול לייצר (עד 10 טוקנים).
# temperature=0.7 → רמת היצירתיות (0–1).
# response.choices[0].text → המענה הראשון שהמודל ייצר.
# .strip() → מנקה רווחים מיותרים


def generate_text(prompt):
    response = openai.Completion.create(
        engine="davinci-002",
        prompt=prompt, # הטקסט שהמודל ימשיך
        max_tokens=10, # כמה טוקנים המודל יחזיר
        temperature=0.7)
    return response.choices[0].text.strip()
```

```python
prompt = "Once upon a time"
```

```python
generated_text = generate_text(prompt)
print(prompt, generated_text)
```

## Customizing the Output

```python
def generate_text(prompt, max_tokens, temperature):
    response = openai.Completion.create(
        engine="davinci-002",
        prompt=prompt, # הטקסט שהמודל ימשיך
        max_tokens=max_tokens, # כמה טוקנים המודל יחזיר
        temperature=temperature)
    return response.choices[0].text.strip()
```

```python
generated_text = generate_text(prompt, 20, 0)
print(prompt, generated_text)
```

```python
generated_text = generate_text(prompt, 20, 1)
print(prompt, generated_text)
```

## Poetic Chatbot

```python
def poetic_chatbot(prompt):
    response = openai.ChatCompletion.create(
        model = "gpt-3.5-turbo",
        messages = [
            {
                "role": "system",
                "content": "You are a poetic chatbot."
            },
            {
                "role": "user",
                "content": "When was Google founded?"
            },
            {
                "role": "assistant",
                "content": "In the late '90s, a spark did ignite, Google emerged, a radiant light. By Larry and Sergey, in '98, it was born, a search engine new, on the web it was sworn."
            },
            {
                "role": "user",
                "content": "Which country has the youngest president?"
            },
            {
                "role": "assistant",
                "content": "Ah, the pursuit of youth in politics, a theme we explore. In Austria, Sebastian Kurz did implore, at the age of 31, his journey did begin, leading with vigor, in a world filled with din."
            },
            {
                "role": "user",
                "content": prompt
            }
        ],
        temperature = 1,
        max_tokens=256
    )
    return response.choices[0].message.content.strip()
```

```python
prompt = "When was cheese first made?"
poetic_chatbot(prompt)
```

```python
prompt = "What will I learn in the Intro to AI course on the 365DataScience platform?"
poetic_chatbot(prompt)
```

## Langchain

```python
from langchain.document_loaders import WebBaseLoader # Loads documents directly from web URLs
from langchain.text_splitter import RecursiveCharacterTextSplitter # Splits long text into smaller, manageable chunks
from langchain.embeddings import OpenAIEmbeddings # Converts text chunks into numerical embeddings using OpenAI
from langchain.vectorstores import FAISS # Stores embeddings in a local vector database for semantic search
from langchain.memory import ConversationBufferMemory # Keeps conversation history to maintain context
from langchain.llms import OpenAI # Provides access to OpenAI text-based LLMs
from langchain.chains import ConversationalRetrievalChain # Combines chat, document retrieval, and LLM responses (RAG)
from langchain.chat_models import ChatOpenAI # Provides access to OpenAI chat-based models (e.g., GPT)
```

```python
url = "https://365datascience.com/courses/intro-to-ai"
```

```python
# טוען תוכן טקסטואלי מאתר אינטרנט והופך אותו ל־ Documents.
loader = WebBaseLoader(url)
```

```python
# Loads the web page content and converts it into Document objects
raw_documents = loader.load() # ממירים אותו ל־ Document objects של LangChain
```

```python
# חלוקת הטקסט ל־ chunks
text_splitter = RecursiveCharacterTextSplitter()
documents = text_splitter.split_documents(raw_documents)
```

```python
# Create an OpenAI embeddings object to convert text into vector representations.
# The API key is sent so OpenAI can authenticate and allow access to their models.
embeddings = OpenAIEmbeddings(openai_api_key = api_key)
```

```python
## Creates embeddings for each text chunk using OpenAI and stores them in a FAISS vector database,
#   enabling fast semantic search over the documents.
vectorstore = FAISS.from_documents(documents, embeddings)
```

```python
# ConversationBufferMemory – stores the conversation
# memory_key – the name the Chain uses for the memory
# return_messages=True – keeps messages in chat format

memory = ConversationBufferMemory(memory_key = "chat_history", return_messages=True)
```

```python
#This is the heart of the system. RAG שרשרת
## The chain does the following:
#1) Receives a question
#2) Searches FAISS for relevant chunks
#3) Sends them to GPT
#4) GPT generates an answer based only on that information

qa = ConversationalRetrievalChain.from_llm(ChatOpenAI(openai_api_key=api_key, 
                                                  model="gpt-3.5-turbo", 
                                                  temperature=0), 
                                           vectorstore.as_retriever(), # מאפשר לשרשרת להשתמש ב־ FAISS כמערכת חיפוש.
                                           memory=memory)
```

```python
query = "What will I learn in the Intro to AI course on the 365DataScience platform?"
```

```python
result = qa({"question": query})
```
